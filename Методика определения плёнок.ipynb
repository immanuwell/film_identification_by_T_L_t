# Строим графики для разных длин волн λ

```python
import pandas as pd

def dict_from_sheet(sheet_name):
	df = pd.read_excel("углерод - для Python.xlsx", skiprows=1, sheet_name=sheet_name)
	df_non_empty = df.dropna(how='all')

	length_time_T = dict()

	for i in range(0, df.dropna(how='all').shape[0]):
	    line = dict(df.iloc[i])
	    length = line["L"]
	    list_D = []
	    for key in line:
	        if "D" in key and len(key)<=3:
	            list_D.append(line[key])

	    list_T = [10**(-D) for D in list_D]
	    list_time = list(range(0, 20, 2))[:len(list_T)]

	    length_time_T[length] = [list_time, list_T]

	    if length <= 300:
	        break

	return length_time_T

all_length_time_T = dict()
for sheet_name in list(pd.ExcelFile("углерод - для Python.xlsx").sheet_names)[:5]:
	all_length_time_T[sheet_name] = dict_from_sheet(sheet_name)
```

```python
import numpy as np

def fit_plane(points):
    # Массив точек
    points = np.array(points)
    
    # Средние значения координат точек
    centroid = points.mean(axis=0)
    
    # Центрируем точки
    centered_points = points - centroid
    
    # Проводим SVD-разложение
    _, _, vh = np.linalg.svd(centered_points)
    
    # Нормаль плоскости будет последней строкой в vh
    normal = vh[-1]
    
    # Уравнение плоскости: ax + by + cz + d = 0
    # где (a, b, c) - нормаль, (x, y, z) - координаты точек на плоскости
    # d = -np.dot(normal, centroid)
    d = -centroid.dot(normal)
    
    return normal, d

def distance_from_plane(point, normal, d):
    # Расстояние от точки до плоскости
    numerator = abs(np.dot(normal, point) + d)
    denominator = np.linalg.norm(normal)
    return numerator / denominator

def closest_surface(point, sheet_name_xyz):
    min_distance = float('inf')
    closest_surface = None
    
    for surface, points in sheet_name_xyz.items():
        normal, d = fit_plane(points)
        distance = distance_from_plane(point, normal, d)
        if distance < min_distance:
            min_distance = distance
            closest_surface = surface
    return closest_surface

# Точка для которой нужно определить ближайшую плоскость
# X-время отжига, Y-пропускание T, Z-длина волны 
point = (6, 0.75, 830)  # a-C H до т-о, правильно
point = (6, 0.84, 830)  # лцу 300, неправильно, tc 300
point = (2, 0.85, 830)  # ta-C 200 имп до т-о, правильно
point = (12, 0.85, 830)  # tc 300, неправильно, a-C H до т-о
point = (10, 0.82, 830)  # a-C H до т-о, неправильно, tc 300

# Нахождение ближайшей плоскости
closest = closest_surface(point, sheet_name_xyz)
print(f"Ближайшая к точке {point} плоскость — это {closest}")
```

# Простой поиск ближайшей поверхности

```python
import numpy as np

def euclidean_distance(point1, point2):
    return np.linalg.norm(np.array(point1) - np.array(point2))

def find_closest_surfaces(point, sheet_name_xyz):
    min_distance = float('inf')
    closest_surfaces = []
    
    for surface, points in sheet_name_xyz.items():
        for surface_point in points:
            distance = euclidean_distance(point, surface_point)
            if distance < min_distance:
                min_distance = distance
                closest_surfaces = [surface]
            elif distance == min_distance:
                if surface not in closest_surfaces:
                    closest_surfaces.append(surface)
    
    return closest_surfaces

# Точка для которой нужно определить ближайшую поверхность
point = (7, 0.72, 830)  # ❶: a-C H до т-о
point = (7, 0.82, 830)  # ❷: tc 300
point = (3, 0.88, 830)  # ❸: ta-C 200 имп до т-о
point = (2, 0.75, 830)  # ❹: a-C H до т-о

closest_surfaces = find_closest_surfaces(point, sheet_name_xyz)
print(f"Данные {point} больше всего соответствуют плёнке: {', '.join(closest_surfaces)}")

# Данные (2, 0.75, 830) больше всего соответствуют плёнке: tc 300
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Преобразование данных в формат (координаты, метка поверхности)
data = []
labels = []
label_map = {}
label_counter = 0

for surface, points in sheet_name_xyz.items():
    if surface not in label_map:
        label_map[surface] = label_counter
        label_counter += 1
    for point in points:
        data.append(point)
        labels.append(label_map[surface])

data = np.array(data)
labels = np.array(labels)

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Создание модели нейросети
model = Sequential()
model.add(Dense(64, input_dim=3, activation='relu'))  # Входной слой (3 координаты)
model.add(Dense(64, activation='relu'))  # Скрытый слой
model.add(Dense(len(label_map), activation='softmax'))  # Выходной слой

# Компиляция модели
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Обучение модели
model.fit(X_train, y_train, epochs=350, batch_size=8, validation_split=0.2)

# Оценка модели на тестовых данных
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test accuracy: {accuracy:.4f}')

# Функция для предсказания поверхности по координатам
def predict_surface(model, point):
    prediction = model.predict(np.array([point]))
    surface_index = np.argmax(prediction)
    for key, value in label_map.items():
        if value == surface_index:
            return key
```
